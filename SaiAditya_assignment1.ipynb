```python
#used a similar logic as hashing
def mode(numbers):
    freq = {}

    for num in numbers:
        if num in freq:
            freq[num] += 1
        else:
            freq[num] = 1

    count = 0
    mode = None

    for num in freq:
        if freq[num] > count:
            count = freq[num]
            mode = num

    return mode

numbers = [1,2,3,5,5,4,6,7,10,10,5]
print(mode(numbers))
```

```python
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('company_sales_data.csv')
data.head(5)
```

```python
plt.plot(data['month_number'],data['total_profit'], '-o')
plt.xlabel("Month Number")
plt.ylabel("Total Profit")
```

```python
month_numbers = data['month_number']
facecream_sales = data['facecream']
facewash_sales = data['facewash']
bar_width = 0.4
position = np.arange(len(month_numbers))
plt.bar(position, facecream_sales, bar_width, label='Face Cream Sales')
plt.bar(position + bar_width, facewash_sales, bar_width, label='Face Wash Sales')
plt.xlabel('Month Number')
plt.ylabel("Sales Units")
plt.legend()
```
